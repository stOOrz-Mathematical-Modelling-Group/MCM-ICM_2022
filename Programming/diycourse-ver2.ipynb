```python
import math
route=400+math.pi*(30+30+30)+400+400+400+math.pi*180/2+2*(180*math.sqrt(3)-180)/2+400
```

```python
print(route)
k=y=[0 for i in range(int(route*10))]
```

```python
for x in range(int(route*10)):
    xx=float(x)
    if xx/10<400:
        k[x]=math.tan(6*math.pi/180)
        y[x]=k[x]*(xx/10)
    elif x<=4000:
        k[x]=math.tan(6*math.pi/180)/2
        y[x]=k[x]*(xx/5)
    elif xx/10<400+math.pi*30:
        k[x]=0
        y[x]=y[4000]
    elif xx/10<400+math.pi*30+200:
        k[x]=-math.tan(6*math.pi/180)
        y[x]=y[4000]+k[x]*(xx/10-math.pi*30-400)
    elif xx/10<600+math.pi*60:
        k[x]=0
        y[x]=y[2000]
    elif xx/10<800+math.pi*60:
        k[x]=math.tan(6*math.pi/180)
        y[x]=y[2000]+k[x]*(xx/10-math.pi*60-600)
    elif xx/10<800+math.pi*90:
        k[x]=0
        y[x]=y[4000]
    elif xx/10<1200+math.pi*90:
        k[x]=-math.tan(6*math.pi/180)
        y[x]=y[4000]+k[x]*(xx/10-math.pi*90-800)
    elif xx/10<1600+math.pi*90:
        k[x]=math.tan(6*math.pi/180)
        y[x]=k[x]*(xx/10-math.pi*90-1200)
    elif xx/10<1600+math.pi*90+math.pi*180/2+(180*math.sqrt(3)-180):
        k[x]=0
        y[x]=y[4000]
    else:
#        print('--')
        k[x]=-math.tan(6*math.pi/180)
        y[x]=k[x]*(xx/10-route)
```

```python
for x in range(int(route*10)):
    print(y[x])
```

```python
import matplotlib.pyplot as plt
dx=dy=[0 for i in range(int(route*10))]
for i in range(int(route*10)):
    dx[i]=i
plt.plot(dx,y)
plt.show()
```
