```python
import math
import numpy as np
import matplotlib.pyplot as plt
k5=2403.5/2
k6=k5/841
x=[0 for i in range(6000)]
y=[0 for i in range(len(x))]
for i in range(3101):
    x[i]=0.01*i
    y[i]=k5*(math.sin((x[i]-15.5)*math.pi/31)+1)
for i in range(3101,6000):
    x[i]=0.01*i
    y[i]=-k6*(x[i]-31)*(x[i]-31)+2*k5


fig = plt.figure(figsize = (8,5), dpi = 800)
plt.title('The energy function with aging')
plt.xlabel('Age(years)')
plt.ylabel('Energy(J)')
plt.plot(x,y)
plt.legend(['A(x)'],loc=4)
plt.show()
```
