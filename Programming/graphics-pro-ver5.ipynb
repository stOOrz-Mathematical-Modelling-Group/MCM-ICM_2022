```python
from PIL import Image
import numpy as np
im=Image.open(r"d:/mathmodelling/2022mcm-icm/tokyo--course.jpg")
maxns=[0 for i in range(10000)]
for x in range(im.size[0]):
    tot=0
#    print(x)
    for y in range(im.size[1]):
        pix=im.getpixel((x,y))
        if(pix[2]<=120):
            tot+=1
        else:
            maxns[x]=max(maxns[x],tot)
            tot=0
        pre=pix
        maxns[x]=max(maxns[x],tot)
# print(maxns[sz-50:sz])
import matplotlib.pyplot as plt
datax = datay = [0 for i in range(im.size[0])]
datax=range(im.size[0])
for i in range(im.size[0]):
    datay[i]=maxns[datax[i]]
sz = im.size[0]
new_data = [0 for i in range(sz)]

stretch = sz/52.8
newdatax = [(i-11)/stretch-4.4 for i in range(sz)]
newdatay = [datay[i]*52.5/83-4.4 for i in range(sz)]

fig=plt.figure(figsize=(10,2),dpi=80)
plt.plot(newdatax[71:sz-45],newdatay[71:sz-45])
plt.ylim(0,70)
plt.xlim(0,44.2)
plt.xlabel('distance from the start(km)')
plt.ylabel('elevation(m)')
plt.show()
```
