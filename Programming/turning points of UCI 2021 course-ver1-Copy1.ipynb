```python
def getsg(x1,x2,y1,y2):
    dtx=[0 for i in range(1000)]
    dty=[0 for i in range(1000)]
    k=(y2-y1)/(x2-x1)
    b=y1-k*x1
    i=0
    while i<=int(x2*100-x1*100):
        dtx[i]=x1+0.01*i
        dty[i]=k*dtx[i]+b
        i+=1
    plt.plot(dtx[0:i-1],dty[0:i-1])
#    plt.show()
getsg(0,6,0,6)
```
